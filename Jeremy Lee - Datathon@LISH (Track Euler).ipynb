```python
# Dropping columns

def drop_features(df):
  drop_list = ['id', 'recorded_by', 'extraction_type', 'extraction_type_group',
               'management', 'payment', 'water_quality', 'quantity_group',
               'source', 'source_class', 'waterpoint_type', 'region', 'date_recorded']
  return df.drop(columns=drop_list)

X_train = drop_features(X_train)
```

```python
# Dealing with missing values
def deal_missing(df):
  fill_mode_cols = ['permit', 'public_meeting', 'scheme_management', 'scheme_name']
  fill_missing = ['subvillage', 'funder', 'installer']

  values = {}
  for col in fill_mode_cols:
    values[col] = df[col].mode()[0]

  for col in fill_missing:
    values[col] = "Unknown"

  filled_df = df.fillna(value=values)
  return filled_df, values

X_train, impute_vals = deal_missing(X_train)
```

```python
def fill_zero_vals(df):
  cols = ['population', 'amount_tsh']
  media_dict = {}
  for col in cols:
    media = df.loc[df[col] != 0, col].median()
    df.loc[df[col] == 0, col] = media
    media_dict[col] = media
  return df, media_dict

def apply_fill_zero_vals(df, media_dict):
  for col in media_dict.keys():
    df.loc[df[col] == 0, col] = media_dict[col]
  return df

X_train, media_dict = fill_zero_vals(X_train)
```

```python
# Get distance from centroid of latitude and longitude
def get_centroid(df):
  return (df.latitude.mean(), df.longitude.mean())

def get_distance(df, centroid):
  df["distance_to_centroid"] = np.sqrt((df['latitude'] - centroid[0])**2 + (df['longitude'] - centroid[1])**2)
  return df

centroid = get_centroid(X_train)
X_train = get_distance(X_train, centroid)
```

```python
# Bin construction year
mean_year = X_train.loc[X_train['construction_year'] != 0, 'construction_year'].median()
def bin_construction(year):
  if year == 0:
    return str(mean_year // 10) + "0s"
  return str(year // 10) + "0s"

def bin_year(df):
  df['construction_year'] = df['construction_year'].apply(bin_construction)
  return df

X_train = bin_year(X_train)
```

```python
# Standardizing numerical columns
def get_scalers(df):
  numerical_cols = ['population', 'amount_tsh', 'gps_height', 'latitude', 'longitude', 'distance_to_centroid']
  scalers = {}

  for col in numerical_cols:
    new_scaler = MinMaxScaler()
    new_scaler.fit(df[col].values.reshape(-1, 1))
    scalers[col] = new_scaler

  return scalers

def standardize(df, scalers):
  for col in scalers.keys():
    df[col] = scalers[col].transform(df[col].values.reshape(-1, 1))
  
  return df

scalers = get_scalers(X_train)
X_train = standardize(X_train, scalers)
```

```python
# Extracting large and small categories
def get_categories(df, threshold=10):
  large_cats = []
  small_cats = []
  for col in df.columns:
    if df[col].dtype == 'O' or df[col].dtype == np.bool or col in ["region_code", "district_code", "month_recorded"]:
      if len(df[col].unique()) > threshold:
        large_cats.append(col)
      else:
        small_cats.append(col)

  return large_cats, small_cats

large_cats, small_cats = get_categories(X_train)
```

```python
# Collapsing large categories
def collapse_categories(df, categories, cut_off=0.975):
  category_cols = categories

  for col in category_cols:
    freq_cap = round(cut_off * df[col].shape[0])
    vc = df[col].value_counts()
    keep_items = []
    running_freq = 0
    for item, freq in vc.iteritems():
      keep_items.append(item)
      running_freq += freq
      if running_freq > freq_cap:
        break
    df.loc[~df[col].isin(keep_items), col] = "Other"

  return df

X_train = collapse_categories(X_train, large_cats)
```

```python
# Target encoding large categories
def get_target_encoder(df, categories):
  encoders = {}
  for cat in categories:
    encoder = TargetEncoder()
    encoder.fit(df[cat], y_train)
    encoders[cat] = encoder
  return encoders

def target_encode(df, encoders):
  for col in encoders.keys():
    df[col] = encoders[col].transform(df[col])
  return df

target_encoders = get_target_encoder(X_train, large_cats)
X_train = target_encode(X_train, target_encoders)
```

```python
# One-hot encoding small categories

def one_hot_encode(df, categories):
  for cat in categories:
    dummies = pd.get_dummies(df[cat], prefix=cat)
    df = pd.concat([df, dummies], axis=1)
  df.drop(columns=categories, inplace=True)

  return df

X_train = one_hot_encode(X_train, small_cats)
```

```python
# General formatting function

def feature_engineering(df, pca_on=True):
  df["date_recorded"] = pd.to_datetime(df['date_recorded'])
  df = drop_features(df)
  df = df.fillna(value=impute_vals)
  df['gps_height'] = abs(df['gps_height'])
  df = get_distance(df, centroid)
  df = apply_fill_zero_vals(df, media_dict)
  #df = cap_outliers(df, outlier_vals)
  #df = get_age(df)
  df = bin_year(df)
  #df = get_time_recorded(df)
  #df = apply_impute_amount(df, amount_impute_vals)
  df = log_transform(df)
  df = standardize(df, scalers)
  df = collapse_categories(df, large_cats)
  df = target_encode(df, target_encoders)
  df = one_hot_encode(df, small_cats)

  for col in X_train.columns:
    if col not in df.columns:
      df[col] = 0

  for col in df.columns:
    if col not in X_train.columns:
      df.drop(columns=col, inplace=True)

  if not pca_on:
    return df

  principal_df = pd.DataFrame(pca.transform(df))

  return principal_df

pca_X_test = feature_engineering(X_test)
```

```python
# Evaluating function

def evaluate_model(model, pred_f='predict_proba', plot=True):
  if pred_f == 'predict_proba':
    predictions = model.predict_proba(pca_X_test)
  elif pred_f == 'decision_function':
    d = ridge.decision_function(pca_X_test)
    predictions = np.exp(d) / (1 + np.exp(d))
  clipped_pred = np.clip(predictions, 0.025, 0.975)
  print(log_loss(y_true=y_test, y_pred=clipped_pred))
  if plot:
    bin_predictions = model.predict(pca_X_test)
    fig = ConfusionMatrixDisplay(confusion_matrix(y_true=y_test, y_pred=bin_predictions))
    fig.plot()

  return predictions
```
